```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as py
```

```python
Dataset=pd.read_csv("Social_Network_Ads.csv")
```

```python
Dataset
```

```python
Dataset=Dataset.drop("User ID",axis=1)
```

```python
Dataset
```

```python
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
```

```python
Dataset
```

```python
Dataset["Purchased"].value_counts()
```

```python
Dataset.columns
```

```python
independent=Dataset[['Age', 'EstimatedSalary', 'Gender_Male']]
independent
```

```python
Dependent=Dataset[["Purchased"]]
```

```python
Dependent_squeeze=np.squeeze(Dependent)
```

```python
Dependent_squeeze
```

```python
# split test and train data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,Dependent_squeeze,test_size=1/3,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
# model creation
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy',max_features='sqrt',splitter='random',random_state=None,max_depth=6)
classifier.fit(X_train,Y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
# evaluation metrics for classification is confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
```

```python
print(cm)
```

```python
# classification report
from sklearn.metrics import classification_report
clf_report=classification_report(Y_test,y_pred)
```

```python
print(clf_report)
```
