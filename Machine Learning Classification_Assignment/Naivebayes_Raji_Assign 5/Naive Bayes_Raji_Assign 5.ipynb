```python
import numpy as np
import pandas as pd
```

```python
Dataset=pd.read_csv("CKD.csv")
```

```python
Dataset
```

```python
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
```

```python
Dataset
```

```python
Dataset.columns
```

```python
# input and output data
independent=Dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
independent
```

```python
dependent=Dataset["classification_yes"]
```

```python
dependent
```

```python
dependent.value_counts()
```

```python
#train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
# preproncessing - Standard scaler
from sklearn.preprocessing import StandardScaler
Sc=StandardScaler()
X_train=Sc.fit_transform(X_train)
X_test=Sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
# model selection
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
param_grid={'var_smoothing':np.logspace(0,-9,num=100)}
grid=GridSearchCV(GaussianNB(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
re=grid.cv_results_
```

```python
y_predict=grid.predict(X_test)
```

```python
grid.predict_proba(X_test)[:,1]
```

```python
# evaluation metrics
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_predict)
print("confusion matrix:\n",cm)
from sklearn.metrics import classification_report
clf_rpt=classification_report(Y_test,y_predict)
print("classification report:\n",clf_rpt)
from sklearn.metrics import f1_score
f1_macro=f1_score(Y_test,y_predict)
print("The f1_macro for best parameter{}".format(grid.best_params_),"\n",f1_macro)
from sklearn.metrics import roc_auc_score
roc_score=roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
print("roc_auc_score:\n",roc_score)


```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
table.to_excel("NaiveBayes_GaussianNB_CKD.xlsx",index=False)
```
