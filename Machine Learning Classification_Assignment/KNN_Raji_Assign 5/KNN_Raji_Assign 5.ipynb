```python
import numpy as np
import pandas as pd
```

```python
Dataset=pd.read_csv("CKD.csv")
```

```python
Dataset
```

```python
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
```

```python
Dataset
```

```python
Dataset.columns
```

```python
# input and output data
independent=Dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python
dependent=Dataset["classification_yes"]
dependent
```

```python
# train and test data split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
Sc=StandardScaler()
X_train=Sc.fit_transform(X_train)
X_test=Sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
param_grids={'n_neighbors':[7,5],
             'weights':['uniform','distance'],
             'algorithm':['auto','ball_tree','kd_tree','brute'],
             'p':[2,1],
             'metric':['minkowski']}
grid=GridSearchCV(KNeighborsClassifier(),param_grids,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
# fitting the model
grid.fit(X_train,Y_train)
```

```python
y_predict=grid.predict(X_test)
```

```python
re=grid.cv_results_
```

```python
#metrics evaluation
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_predict)
print("Confusion matrix:\n",cm)
```

```python
from sklearn.metrics import classification_report
clf_rpt=classification_report(Y_test,y_predict)
print("Classsification_report:\n",clf_rpt)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(Y_test,y_predict)
print("The f1_macro value for best parameter:{}".format(grid.best_params_),"\n",f1_macro)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```
